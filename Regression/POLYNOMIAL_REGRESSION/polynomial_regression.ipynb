```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d=pd.read_csv('/home/bhanu/Documents/ML/Regression/POLYNOMIAL_REGRESSION/Position_Salaries.csv')
X=d.iloc[:,1:-1].values
y=d.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
linear_regression=LinearRegression()
linear_regression.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
a=PolynomialFeatures(degree=20)
X_ploy=a.fit_transform(X)
lin_reg_2=LinearRegression()
lin_reg_2.fit(X_ploy,y)

```

```python
plt.scatter(X,y,color='red')
plt.plot(X,linear_regression.predict(X),color='blue')
plt.title("tor b")
plt.xlabel('psoitionlevel')
plt.ylabel('slary')
plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(X_ploy),color='blue')
plt.title("tor b")
plt.xlabel('psoitionlevel')
plt.ylabel('slary')
plt.show()
```
